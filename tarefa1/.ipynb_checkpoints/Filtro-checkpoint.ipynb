```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
imagem = "500.png"

img = Image.open(imagem)
img = img.convert('L') #Convert to gray scale
#img = img.resize((IMG_SIZE, IMG_SIZE), Image.ANTIALIAS)
print("Dimencoes da imagem = {}".format(img.size))
IMG_SIZE_x = img.size[1]
IMG_SIZE_y = img.size[0]
```

```python
plt.imshow(img)
```

```python
cores = np.array(img)
```

```python
def smoothing(imagem, i, j):
    limite_y, limite_x = imagem.shape
    if ( (j == limite_x-1) or (i == limite_y-1) or (i == 0) or (j == 0)) :
        return imagem[i][j]
    else:
        return ( imagem[i+1][j] + imagem[i][j+1] + imagem[i-1][j] + imagem[i][j-1] ) / 4
```

```python
def derivadaX(imagem, i, j):
    return abs( 2*imagem[i][j] -imagem[i][j-1] - imagem[i][j+1] )
    return abs( 6*imagem[i][j] -2*imagem[i][j-1] -2*imagem[i][j+1] -imagem[i][j-2] -imagem[i][j+2] )
    return 255 if 6*imagem[i][j] -2*imagem[i][j-1] -2*imagem[i][j+1] -imagem[i][j-2] -imagem[i][j+2] >= 50 else 0
def derivadaY(imagem, i, j):
    return abs( 2*imagem[i][j] -imagem[i-1][j] - imagem[i+1][j] )
    return 255 if 2*imagem[i][j] -imagem[i-1][j] - imagem[i+1][j] >= 50 else 0
```

```python
filtro_smooth = []

for i in range(IMG_SIZE_x):
    filtro_smooth.append([])
    for j in range(IMG_SIZE_y):
#        print("{} {}".format(i,j))
        filtro_smooth[-1].append( smoothing(cores,i, j) )
```

```python
filtro = []
cores = filtro_smooth
for i in range(1,IMG_SIZE_x-1):
    filtro.append([])
    for j in range(1, IMG_SIZE_y-1):
        filtro[-1].append( derivadaX(cores, i, j) + derivadaY(cores, i, j) )
```

```python
fig, axs = plt.subplots(1, 2, figsize=(9, 3), sharey=True)

axs[0].imshow( filtro ,cmap='Greys')
axs[1].imshow( cores )
```
