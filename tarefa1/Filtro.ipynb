```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
IMG_SIZE = 300
```

```python
imagem = "dados.jpeg"
img = Image.open(imagem)
img = img.convert('L')
#img = img.resize((IMG_SIZE, IMG_SIZE), Image.ANTIALIAS)
print(img.size)
IMG_SIZE_x = img.size[1]
IMG_SIZE_y = img.size[0]
```

```python
plt.imshow(img)
```

```python
cores = np.array(img)
```

```python
def derivadaX(imagem, i, j):
    return 255 if 6*imagem[i][j] -2*imagem[i][j-1] -2*imagem[i][j+1] -imagem[i][j-2] -imagem[i][j+2] >= 50 else 0
def derivadaY(imagem, i, j):
    return 255 if 2*imagem[i][j] -imagem[i-1][j] - imagem[i+1][j] >= 50 else 0
```

```python
filtro = []
for i in range(2,IMG_SIZE_x-2):
    filtro.append([])
    for j in range(2, IMG_SIZE_y-2):
        filtro[-1].append( derivadaX(cores, i, j) + derivadaY(cores, i, j) )
```

```python
fig, axs = plt.subplots(1, 2, figsize=(9, 3), sharey=True)

axs[0].imshow( filtro ,cmap='Greys')
axs[1].imshow( img ,cmap='Greys')

plt.plot()
```
