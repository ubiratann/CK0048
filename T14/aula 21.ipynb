```python
def print_matriz(matriz, erro=10**-9):
    n = len(matriz)
    for i in range(n):
        for j in range(n):
            numero = float(matriz[i][j])
            if numero<=erro:
                numero = float(0)
            print("{:.3}".format(numero), end="\t")
        print()

def identidade(n):
    identidade = [[0 for j in range(n)] for i in range(n)]
    for i in range(n):
        identidade[i][i] = 1
    return identidade

def multiplica(m1, m2):
    n = len(m1)
    if n != len(m2):
        print("erro dimensão")
        return None
    result = identidade(n)
    for i in range(n):
        for k in range(n):
            somatorio = 0
            for j in range(n):
                somatorio += m1[i][j] * m2[j][k]
            result[i][k] = somatorio
    return result
        
                
def Transposta(matriz):
    n = len(matriz)
    return [ [ matriz[i][j] for i in range(n)] for j in range(n)]

A = [
    [40, 8, 4, 2, 1],
    [8, 30, 12, 6, 2],
    [4, 12, 20, 1, 2],
    [2, 6, 1, 25, 4],
    [1, 2, 2, 4, 5]]
teste = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]]
```

```python
def somaDosQuadradosDosTermosAbaixoDaDiagonal(matriz):
    somatorio = 0
    n = len(matriz)
    for i in range(n):
        for j in range(n):
            if i>j:
                somatorio += math.pow(matriz[i][j],2)
    return somatorio
```

```python
def matrizJacobiBaseadaNoElemento_ij_DaMatrizVelha(matriz,i,j):
    pi = math.pi
    n = len(matriz)
    J = identidade(n)
    erro = 10**(-9)
    teta = 0
    
    if( abs(matriz[i][j]) <= erro ):
        return J
    elif( abs( matriz[i][i] - matriz[j][j] ) <= erro):
        teta = pi/4
    else:
        teta = (1/2) * math.atan( (-2*matriz[i][j]) / (matriz[i][i] - matriz[j][j]) )
    
    J[i][i] = math.cos(teta)
    J[j][j] = math.cos(teta)
    J[i][j] = math.sin(teta)
    J[j][i] = -1 * math.sin(teta)
    
    return J
```

```python
def varreduraDeJacobi(matriz, householder = False):
    n = len(matriz)
    J = identidade(n)
    
    velho_a = matriz
    for j in range(n-1):
        n2 = n if not householder else j+2
        for i in range(j+1, n2):
#             print(i)
            j_matriz = matrizJacobiBaseadaNoElemento_ij_DaMatrizVelha(velho_a, i, j)
            
            j_t = Transposta(j_matriz)
            novo_a = multiplica(j_t, velho_a)
            novo_a = multiplica(novo_a, j_matriz)
            
            velho_a = novo_a
            J = multiplica(J, j_matriz)
            
    print("Varredura")
    print_matriz(novo_a, 10**-3)

    print()
    
    return (novo_a, J)
```

```python
def Algoritmo_Jacobi(matriz_a, erro_aceitavel = 10**-15, householder = False):
    n = len(matriz_a)
    loss = 100
    P = identidade(n)
    
    velho_a = matriz_a
    contador = 0
    while erro_aceitavel <= loss:
        contador += 1
        novo_a, j = varreduraDeJacobi(velho_a, householder)
        velho_a = novo_a
        
        P = multiplica(P,j)
        loss = somaDosQuadradosDosTermosAbaixoDaDiagonal(novo_a)
        if householder and contador == 10:
            break
    return (novo_a, P)
```
