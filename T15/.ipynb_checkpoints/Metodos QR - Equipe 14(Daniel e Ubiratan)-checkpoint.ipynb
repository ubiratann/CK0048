```python
import numpy as np
import math
```

```python
def print_matriz(matriz, erro=10**-9):
    n = len(matriz)
    for i in range(n):
        for j in range(n):
            numero = float(matriz[i][j])
            if numero<=erro:
                numero = float(0)
            print("{:.3}".format(numero), end="\t")
        print()

def identidade(n):
    identidade = [[0 for j in range(n)] for i in range(n)]
    for i in range(n):
        identidade[i][i] = 1
    return identidade

def multiplica(m1, m2):
    n = len(m1)
    if n != len(m2):
        print("erro dimensão")
        return None
    result = identidade(n)
    for i in range(n):
        for k in range(n):
            somatorio = 0
            for j in range(n):
                somatorio += m1[i][j] * m2[j][k]
            result[i][k] = somatorio
    return result
        
                
def Transposta(matriz):
    n = len(matriz)
    return [ [ matriz[i][j] for i in range(n)] for j in range(n)]

A = [
    [40, 8, 4, 2, 1],
    [8, 30, 12, 6, 2],
    [4, 12, 20, 1, 2],
    [2, 6, 1, 25, 4],
    [1, 2, 2, 4, 5]]
```

```python
def somaDosQuadradosDosTermosAbaixoDaDiagonal(matriz):
    somatorio = 0
    n = len(matriz)
    for i in range(n):
        for j in range(n):
            if i>j:
                somatorio += math.pow(matriz[i][j],2)
    return somatorio
```

```python
def matrizJacobiBaseadaNoElemento_ij_DeRvelha(matriz, i, j):
    pi = math.pi
    n = len(matriz)
    erro = 10**-5
    
    J = identidade(n)
    if abs(matriz[i][j]) < erro :
        return J
    
    if abs(matriz[j][j]) <= erro :
        if matriz[i][j] < 0:
            teta = pi/2
        else:
            teta = -1 * pi/2
    else:
        teta = math.atan( (-1*matriz[i][j]) / (matriz[j][j]) )
    
    J[i][i] = math.cos(teta)
    J[j][j] = math.cos(teta)
    J[i][j] = math.sin(teta)
    J[j][i] = -1 * math.sin(teta)
    
    return J
```

```python
def decomposicaoQR(matriz):
    n = len(matriz)
    QT = identidade(n)
    velho_r = matriz
    
    for j in range(n-1):
        for i in range(j+1, n):
            J = matrizJacobiBaseadaNoElemento_ij_DeRvelha(velho_r, i, j);
            novo_r = multiplica(J, velho_r)
            velho_r = novo_r
            
            QT = multiplica(J,QT)
            
    Q = Transposta(QT)
    R = novo_r
    return (Q,R)
```

```python
def metodoQR(A,erro_aceitavel = 10**-5):
    n = len(A)
    loss = 100
    velho_a = A.copy()
    P = identidade(n)
    
    while loss>=erro_aceitavel :
        Q, R = decomposicaoQR(velho_a);
        
        novo_a = multiplica(R,Q)
        velho_a = novo_a
        
        P = multiplica( P, Q)
        
        print_matriz(novo_a, 10**-5)
        print()
        
        loss = somaDosQuadradosDosTermosAbaixoDaDiagonal(novo_a)
    return (novo_a, P)
```
